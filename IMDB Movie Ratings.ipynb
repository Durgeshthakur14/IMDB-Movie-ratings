```python
drop_column = df2.drop(["id","release_date","launch Month"],axis=1)
```

#Identify movies with a high popularity score and a high IMDB rating

```python
popularity_threshold = 8.5
rating_threshold = 8.5

popular_and_highly_rated_movies = drop_column[(drop_column['popularity'] >= popularity_threshold) & (drop_column['vote_average'] >= rating_threshold)]

print("Movies with a high popularity score and a high IMDB rating:")
print(popular_and_highly_rated_movies)
```

#Analyze the relationship between release month and IMDB rating

```python
avg_rating_by_month = df.groupby('launch Month')['vote_average'].mean()

print("Average IMDB rating for movies released in different months:")
print(avg_rating_by_month)
```

#Identify movies with a high number of votes and a high IMDB rating


```python
vote_count_threshold = 10000

movies_with_high_votes_and_rating = df[df['vote_count'] >= vote_count_threshold]

avg_rating_for_high_vote_movies = movies_with_high_votes_and_rating['vote_average'].mean()

print("Average IMDB rating for movies with a high number of votes:")
print(avg_rating_for_high_vote_movies)
```

#Analyze the average IMDB rating for movies released in different decades

```python
avg_rating_by_decade = df.groupby('release_date')['vote_average'].mean()

print("Average IMDB rating for movies released in different decades:")
print(avg_rating_by_decade)
```
